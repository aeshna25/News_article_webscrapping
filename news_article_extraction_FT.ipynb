```python
from bs4 import BeautifulSoup
import requests
```

```python
def get_articles(company: str) -> dict:
    # url = 'https://www.ft.com/search?q=tata&dateRange=now-1y&sort=relevance&isFirstView=false' 
    url = f'https://www.ft.com/search?q={company}&dateRange=now-1y&sort=relevance&isFirstView=false' #last 1 year news articles
    headers = {
    "User-Agent": "Mozilla/5.0 (iPhone; CPU iPhone OS 5_1 like Mac OS X) AppleWebKit/534.46 (KHTML, like Gecko) Version/5.1 Mobile/9B179 Safari/7534.48.3"
    }
    response = requests.get(url, headers=headers)

    
    soup = BeautifulSoup(response.content, 'html.parser')
    soup.encode('utf-8')
    # print(response.content)
    urls = soup.find_all(class_ = 'o-teaser__heading') 
    desc = soup.find_all(class_ = 'o-teaser__standfirst') 
    dates = soup.find_all(class_ = 'o-teaser__timestamp') 
    article_name= []
    article_desp =[]
    article_date =[]
    link_list =[]
    for url in urls:
        # print(url)
        name=  url.find('a').text
        article_name.append(name)
    for des in desc:
        desp = des.get_text() 
        # print(desp)
        article_desp.append(desp)
    for date in dates:
        dat = date.get_text() 
        # print(dat)
        article_date.append(dat)
    for link in urls:
        lst = link.find("a")['href']
        # print(lst)
        link_list.append(lst)
    merged = [
        {"Article name": article_name,'Article date': article_date, "Article description": article_desp, "Article link": link_list} 
        for article_name,article_date, article_desp, link_list in zip(article_name,article_date, article_desp, link_list)
                ] 
    return merged    
```

```python
get_articles('zomato')
```
